### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Statement 3:

    Average salary significantly differs among industries.

Null Hypothesis (H₀)

There is no significant difference in average salary across different industries.

Alternative Hypothesis (H₁)

There is a significant difference in average salary across different industries.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import f_oneway


temp_df = df[
    ['Industry', 'avg_salary']
].dropna()


industry_groups = [
    group['avg_salary'].values
    for name, group in temp_df.groupby('Industry')
]


f_statistic, p_value = f_oneway(
    *industry_groups
)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

I used the ANOVA Test (Analysis of Variance) to obtain the P-value because Industry is a categorical variable and average salary is a numerical variable. ANOVA helps determine whether there is a statistically significant difference in average salary across different industries.

##### Why did you choose the specific statistical test?

The ANOVA Test was selected because the dataset contains one categorical variable (Industry) and one numerical variable (avg_salary). ANOVA is appropriate for determining whether there is a statistically significant difference in the mean salary across multiple industry groups.

````python
replacements = {
    "exp": "experience",
    "yrs": "years",
    "yr": "year",
    "&": "and",
    "mgr": "manager",
    "dev": "developer",
    "eng": "engineer",
    "sr": "senior",
    "jr": "junior",
    "dept": "department",
    "hr": "human resources",
    "tech": "technology",
    "req": "required",
    "edu": "education",
    "btech": "bachelor of technology",
    "cs": "computer science",
    "ml": "machine learning",
    "ai": "artificial intelligence",
    "sql": "structured query language"
}
def rephrase_text(text):
    words = str(text).split()

    updated_words = [
        replacements.get(word, word)
        for word in words
    ]

    return ' '.join(updated_words)

df['Job Description'] = df[
    'Job Description'
].apply(rephrase_text)


df['Job Description'].head()
````
